```python
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.io import wavfile
import IPython
from IPython.display import Audio
plt.style.use('ggplot')

AudioFileName = "ton2.wav"
IPython.display.Audio(AudioFileName)

```

```python
# Load audio File
fs, data = wavfile.read(AudioFileName)
print(AudioFileName+', Sample Rate = '+str(fs)+', Wav Format= '+str(data.dtype))

```

```python
# Plot audio signal
plt.rcParams['figure.figsize'] = (14, 10)# figure size

sp1 = plt.subplot(311)
plt.plot(data[:8000],'b')
plt.ylabel('x (n)')
sp1.set_xlim((0,8000))

sp2 = plt.subplot(312)
plt.plot(data[:1000],'b')
plt.ylabel('x (n)')
sp2.set_xlim((0,1000))

sp3 = plt.subplot(313)
plt.stem(data[:100],'b')
plt.ylabel('x [n]')
sp3.set_xlim((0,100))
plt.suptitle('Different time representations for digital audio signals.',size=16)

plt.tight_layout()
plt.subplots_adjust(top=0.9)
```
